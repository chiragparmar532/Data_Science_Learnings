## We will be performing analysis on the Metropolitan Dataset

## Data Dictionary

### land_area : size in square miles
### total_population : estimated population in thousands
### percent_city :  percent of population in central city/cities
### percent_senior: percent of population <=65
### physicians  : number of professionally active physicians
### hospital_beds: total number of hospital beds
### graduates : percent of adults that finished high school
### work_force: number of person in work force in thousands
### income : total income in 1976 in millions of dollars
### region : geographic region
### crime_rate : ratio of number of serious crimes by total population

## Multiple Plots in 1 Figure i.e., Subplots

```python
fig, ax = plt.subplots()
```

## 1 Row 2 Col 2 Subplots Method 1

```python
plt.figure(figsize=(12,8))

plt.subplot(1,2,1) #row, column and index
plt.plot(df.work_force, df.income, "--ro")
plt.title("Work Force Vs Income")
plt.xlabel("Work Force")
plt.ylabel("Income")

plt.subplot(1,2,2) #row, column and index
plt.plot(df.physicians, df.income, "-.b^")
plt.title("Physicians Vs Income")
plt.xlabel("Physicians")
plt.ylabel("Income")

plt.suptitle("Subplots for Work Force Vs Income"); # Add a centered line or a title for the whole subplot
```

## 1 Row, 2 Column 2 Subplots Method 2

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,8)) #row, column
ax1.plot(df.work_force, df.income, "--ro")
ax1.set_title("Work Force Vs Income")

ax2.plot(df.physicians, df.income, ":b^")
ax2.set_title("Physicians Vs Income")

plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(df.income, df.physicians)

plt.subplot(2,2,2)
plt.plot(df.income, df.hospital_beds)
```
